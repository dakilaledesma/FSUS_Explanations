```python
anno_file = open("annotations.txt")
anno_lines = anno_file.readlines()

anno_dict = {}
for line in anno_lines:
  idx, sp = line.strip().split(",")
  anno_dict[int(idx)] = sp.replace(".", '')
```

```python
print(list(anno_dict.items())[:5])
```

```python
import os
import shutil
from glob import glob

if not os.path.isdir("anno_out"):
  os.makedirs("anno_out")

for im_fn in glob("outs/outs/resolution/600_cnl/fullgrad/*.*"):
  bn = os.path.basename(im_fn)
  cat_idx, gbif_id = bn.split("_")
  fn = f"{anno_dict[int(cat_idx)]}_{gbif_id}"
  print(fn)
  
  shutil.move(im_fn, f"anno_out/{fn}")
```
